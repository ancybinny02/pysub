# Create the data visualization using the matplotlib.


```python
import matplotlib.pyplot as plt
```

## Visualize the Gender of Passengers using the Bar graph.

```python
gender_counts = df["Sex"].value_counts()

plt.bar(gender_counts.index, gender_counts.values)
plt.xlabel("Gender")
plt.ylabel("Number of passengers")
plt.title("Gender of Titanic passengers")
plt.show()
```

## Visualize the Survival Count of Passengers using the Bar graph.

```python
survival_counts = df["Survived"].value_counts()

survival_counts.index = ["Not survived", "Survived"]

plt.bar(survival_counts.index, survival_counts.values)
plt.xlabel("Survival status")
plt.ylabel("Number of passengers")
plt.title("Survival count of Titanic passengers")
plt.show()
```

## Visualize the Age of Passengers using the Bar/Histogram graph.

```python
plt.hist(df["Age"], bins=20)
plt.xlabel("Age")
plt.ylabel("Number of passengers")
plt.title("Age distribution of Titanic passengers")
plt.show()
```

## Visualize the comparison of Age and Fare of Passengers using the Scatterplot.

```python
plt.scatter(df["Age"], df["Fare"])
plt.xlabel("Age")
plt.ylabel("Fare")
plt.title("Age and fare comparison of Titanic passengers")
plt.show()
```
